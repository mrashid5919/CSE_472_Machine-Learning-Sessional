Data cleaning

```python
#replacing missing values with columnwise mean
numeric_cols = dataframe.select_dtypes(include='number')
dataframe[numeric_cols.columns] = numeric_cols.fillna(numeric_cols.mean())
```

```python
#removing duplicates
dataframe.drop_duplicates(inplace=True)
```

```python
#dropping rows with missing value in target column
dataframe.dropna(subset=['Attrition'], inplace=True)
```

Creation of input and output features

```python
features=dataframe.drop('Attrition',axis=1)
target=dataframe['Attrition']
#features
```

Conversion of features into numeric values

```python
#dropping the columns with single distinct value
feature_encoded=features.drop(['EmployeeCount','Over18','StandardHours'],axis=1)
```

```python
#label encoding the columns with 2 distinct values
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
target_encoded=encoder.fit_transform(target)

columns_to_encode = ['Gender','OverTime','PerformanceRating']
#feature_encoded=features.copy()
for cols in columns_to_encode:
    feature_encoded[cols]=encoder.fit_transform(features[cols])
    
#features
```

```python
#one hot encoding the columns with more than 2 distinct values
categorical_columns=[col for col in features if features[col].nunique() > 2 and features[col].dtype == 'object']
#print(categorical_columns)

for col in categorical_columns:
    feature_encoded=pd.get_dummies(feature_encoded, columns=[col], drop_first=True)

#feature_encoded
```

Scaling of the features

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

def scaling(features_encoded,scaling_type):
    not_one_hot_columns=[col for col in features_encoded if features_encoded[col].nunique() > 2]
    #print(not_one_hot_columns)
    if scaling_type=='StandardScaler':
        scaler=StandardScaler()
    elif scaling_type=='MinMaxScaler':
        scaler=MinMaxScaler()
    features_encoded[not_one_hot_columns]=scaler.fit_transform(features_encoded[not_one_hot_columns])
    return features_encoded
```

```python
#scale data
feature_scaled=scaling(feature_encoded,'StandardScaler')
#feature_scaled
```

Correlation

```python
#Correlation analysis features with target
features_df=pd.DataFrame(feature_scaled,columns=feature_encoded.columns)
target_df=pd.DataFrame(target_encoded,columns=['Attrition'])
correlations=features_df.corrwith(target_df['Attrition'])
correlations
```

```python
#top 20 features with highest correlation
top_twenty_features=correlations.abs().sort_values(ascending=False).head(20)
top_twenty_features
```

```python
#1d scatter plots
import matplotlib.pyplot as plt
import numpy as np
for f in top_twenty_features.index:
    class_0=features_df.loc[target_df['Attrition']==0]
    class_1=features_df.loc[target_df['Attrition']==1]

    plt.plot(class_0[f],np.zeros_like(class_0[f]),'o',label='Class 0')
    plt.plot(class_1[f],np.zeros_like(class_1[f]),'o',label='Class 1')

    plt.legend()
    plt.xlabel(f)
    plt.title('1D scatter plot of '+ f+ ' by Numeric Classes')
    plt.show()
```
